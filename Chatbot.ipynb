```python
import io
import os
import re
import hashlib
import faiss
import numpy as np
import getpass
from dataclasses import dataclass
from PyPDF2 import PdfReader
from openai import OpenAI
api_key = getpass.getpass("Enter your OpenAI API key: ")
client = OpenAI(api_key=api_key)
EMBEDDING_MODEL = "text-embedding-3-small"

@dataclass
class Chunk:
    text: str
    source: str
    page: int
    chunk_id: str


def clean_text(text):
    text = text.replace("\n", " ")
    text = re.sub(r"\s+", " ", text)
    return text.strip()


def extract_text_by_page(filename, file_bytes):
    if filename.lower().endswith(".pdf"):
        reader = PdfReader(io.BytesIO(file_bytes))
        pages = []
        for i, page in enumerate(reader.pages, start=1):
            text = page.extract_text() or ""
            text = clean_text(text)
            if text:
                pages.append((i, text))
        return pages
    else:
        text = file_bytes.decode("utf-8", errors="ignore")
        return [(1, clean_text(text))]


def chunk_text_by_page(page_text, source, page, chunk_size=800, overlap=100):
    chunks = []
    start = 0
    while start < len(page_text):
        end = min(start + chunk_size, len(page_text))
        chunk = page_text[start:end]
        chunk_id = hashlib.md5(f"{source}:{page}:{start}:{chunk}".encode()).hexdigest()
        chunks.append(Chunk(chunk, source, page, chunk_id))
        if end == len(page_text):
            break
        start = end - overlap
    return chunks


def embed_texts(texts):
    response = client.embeddings.create(
        model=EMBEDDING_MODEL,
        input=texts
    )
    vectors = [d.embedding for d in response.data]
    arr = np.array(vectors, dtype="float32")
    faiss.normalize_L2(arr)
    return arr


def build_faiss(chunks):
    texts = [c.text for c in chunks]
    embeddings = embed_texts(texts)
    dim = embeddings.shape[1]
    index = faiss.IndexFlatIP(dim)
    index.add(embeddings)
    return index, chunks


def retrieve(query, index, chunks, top_k=3):
    q_vec = embed_texts([query])
    scores, indices = index.search(q_vec, top_k)

    results = []
    for score, idx in zip(scores[0], indices[0]):
        if idx != -1:
            results.append({
                "text": chunks[idx].text,
                "source": chunks[idx].source,
                "page": chunks[idx].page,
                "score": float(score)
            })
    return results


def extract_exact_answer(question, text):
    q = question.lower().strip().rstrip("?")
    text_lower = text.lower()

    patterns = [
        rf"\d+\)\s*{re.escape(q)}\s*\?",
        rf"{re.escape(q)}\s*\?"
    ]

    start = -1
    matched = None

    for pattern in patterns:
        m = re.search(pattern, text_lower)
        if m:
            start = m.start()
            matched = m.group()
            break

    if start == -1:
        return text

    remaining = text[start:]

    next_q = re.search(r"\s\d+\)\s+[A-Za-z].*?\?", remaining[1:])
    if next_q:
        end = next_q.start() + 1
        return remaining[:end].strip()

    return remaining.strip()
def answer_question(question, results):
    if not results:
        return "No relevant content found."

    text = results[0]["text"]

    # remove question part (before ?)
    if "?" in text:
        text = text.split("?", 1)[1]

    # clean broken words (like Strin g → String)
    text = text.replace(" ", " ").strip()

    # stop at next question number (but safely)
    parts = re.split(r"\s\d+\)\s", text)

    return parts[0].strip()
```

```python
all_chunks = []

for filename, file_data in uploaded.items():
    pages = extract_text_by_page(filename, file_data)

    for page_num, page_text in pages:
        chunks = chunk_text_by_page(page_text, source=filename, page=page_num, chunk_size=500, overlap=100)
        all_chunks.extend(chunks)

print(f"Total chunks: {len(all_chunks)}")

index, chunk_store = build_faiss(all_chunks)
```

```python
while True:
    question = input("Ask a question (or type 'exit'): ")

    if question.lower() == "exit":
        break

    results = retrieve(question, index, chunk_store, top_k=3)
    answer = answer_question(question, results)
    print(answer)
```
